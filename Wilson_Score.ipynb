```python
import pandas as pd
from scipy.stats import norm

# Read the CSV file into a pandas dataframe
df = pd.read_csv('/content/drive/My Drive/wine_rev/combined_dataset.csv')

# Group the dataframe by country and calculate the count of each country
country_counts = df.groupby('country')['country'].count()

# Group the dataframe by country and calculate the average point of each country
country_avg_points = df.groupby('country')['points'].mean()


n = country_counts
p = country_avg_points / 100


# Calculated the Wilson score of each country's average points
z = norm.ppf(1-0.05/2)
wilson_score = ((p + z**2/(2*n) - z*((p*(1-p)/n + z**2/(4*n**2))**0.5))/(1+z**2/n)).sort_values(ascending=False)

# Print the results
print(wilson_score*100)
```

```python
import pandas as pd
from scipy.stats import norm

# Read the CSV file into a pandas dataframe
df = pd.read_csv('/content/drive/My Drive/wine_rev/combined_dataset.csv')

# Group the dataframe by country and calculate the count of each country
variety_counts = df.groupby('variety')['variety'].count()


variety_avg_points = df.groupby('variety')['points'].mean()

# Calculate the sample size (n) and sample portion (p) for each country
n = variety_counts
p = variety_avg_points / 100

# Calculate the Wilson score for each country with points > 90
z = norm.ppf(1-0.05/2)
wilson_score = ((p + z**2/(2*n) - z*((p*(1-p)/n + z**2/(4*n**2))**0.5))/(1+z**2/n)).sort_values(ascending=False)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Print the results
print(wilson_score*100)
```
