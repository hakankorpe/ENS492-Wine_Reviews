```python
import pycountry 
def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
           # .alpha_3 means 3-letter country code 
           # .alpha_2 means 2-letter country code
            CODE.append(code)
        except:
            CODE.append('None')
    return CODE
# create a column for code 
merged_dataframes['CODE']=alpha3code(merged_dataframes.country)
merged_dataframes.loc[merged_dataframes.country == 'US', 'CODE'] = 'USA'
merged_dataframes.loc[merged_dataframes.country == 'England', 'CODE'] = 'GBR'
merged_dataframes.loc[merged_dataframes.country == 'Czech Republic', 'CODE'] = 'CZE'
merged_dataframes.loc[merged_dataframes.country == 'Macedonia', 'CODE'] = 'MKD'
merged_dataframes.loc[merged_dataframes.country == 'Moldova', 'CODE'] = 'MDA'

merged_dataframes.head()
```
